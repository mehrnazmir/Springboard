# Find the top 10 major project themes:

```python
#First see the mjtheme_namecode column of json_df dataframe 
json_df.mjtheme_namecode
```

 It is a nested element. Therefore, normalization should be performed.
 As a first step, the json file is loaded as string. Then, use normalization to create tables from nested element. 


```python
#load the json file as string 
with open('/Users/mehrnaz/Desktop/SpringBoard/Assignment/Data_Wrangling/mini_proj1/data_wrangling_json/data/world_bank_projects.json', 'r') as json_file:
    json_data = json.load(json_file)
```

```python
# use normalization to create tables from nested element
normalized_mjtheme_namecode = json_normalize(json_data, 'mjtheme_namecode')
```

```python
#See 10 rows
normalized_mjtheme_namecode.head(10)
```

```python
#There is missing value for name
#find the top 10 from code 
top_most_proj = normalized_mjtheme_namecode.code.value_counts()
top_most_proj
```
